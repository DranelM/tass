```python
import seaborn as sns
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

use_cols = ['user_id', 'review_count', 'friends', 'average_stars']

df = pd.read_csv('yelp_user.csv', usecols=use_cols, dtype={'review_count': np.int16, 'average_stars': np.float16})
```

## Check users without friends

```python
#liczba użytkowników
df.shape[0]
```

```python
# Sprawdzenie liczby osób bez znajomych
df[df['friends'] == 'None'].describe()
```

```python
# ile procent osób nie ma ani jednego znajomego
(df[df['friends'] == 'None'].count() / df.shape[0])[0]
```

<b> Aż 42% użytkowników nie posiada żadnego znajomego. Nie będą oni brani pod uwagę podczas odtwarzania struktur powiązań użytkowników.<br><br>

```python
# usunięcie z dataframe wszystkich użytkowników bez znajomych
df = df[df['friends'] != 'None']
```

```python
df.shape
```

<b> Zostało tylko 760007 użytkowników <br><br>

```python
# relacja między średnią oceną a liczbą wystawionych ocen
sns.scatterplot('average_stars', 'review_count', data=df)
```

## number of Friends

```python
fr = df[['friends']].copy()
df['nr_of_friends'] = fr['friends'].map(lambda x: len(x.split(','))).astype(np.int16)
df.sort_values(['nr_of_friends'], ascending=False).head(100)
```

```python
# relacja między liczbą znajomych a napisanych recenzji
sns.scatterplot('nr_of_friends', 'review_count', data=df);
```

```python
pos = df[df['average_stars'] >= 3].index
neg = df[df['average_stars'] < 3].index
```

```python
print(f'W bazie znajduje się {df.loc[pos].count()[0]} osób nastawionych pozytywnie')
print(f'W bazie znajduje się {df.loc[neg].count()[0]} osób nastawionych negatywnie')
```

```python
review_per_positiveu = df.loc[pos]['review_count'].mean()
review_per_negativeu = df.loc[neg]['review_count'].mean()
pos_to_neg = review_per_positiveu / review_per_negativeu
print("Negative mean review_count : ", review_per_negativeu)
print("Positive mean review_count : ", review_per_positiveu)
print("Pos to neg ratio: ", pos_to_neg)
```

Osoba pozytywna w stosunku do osoby negatywnie nastawionej komentuje 3.49 raza częściej

## Create Network

```python
df[df['average_stars'] >= 3].count()[0] / df.shape[0]
```

```python
df.shape
```

```python
df_connections = pd.concat([pd.Series(row['user_id'], row['friends'].split(',')) for _, row in df.iloc[300000:500000].iterrows()]).reset_index()
```

```python
df_connections.columns = ['user_id', 'friends']
```

```python
df_connections.dtypes
```

```python
df_connections.head()
```

```python
G = nx.from_pandas_edgelist(df_connections, 'user_id', 'friends')
```

```python
print(nx.info(G))
```

```python
print(nx.info(G))
```

```python
df['nr_of_friends'].mean()
```

```python
ego = nx.ego_graph(G, df_connections['friends'][0])
```

```python
plt.figure(figsize=(15,15))
nx.draw(ego, with_labels=True)
plt.show()
```
